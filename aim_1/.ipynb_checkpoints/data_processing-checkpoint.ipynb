# Here we will take the processed data from the pipeline first_pipeline.sh and make some pretty figures

```python
import matplotlib.pyplot as plt
import numpy as np
```

### Plot first graph

```python
data_array = np.array([[1,2,3,4,5],[1,4,9,16,25]])

plt.plot(data_array[0,],data_array[1,])
plt.xlabel('x label')
plt.ylabel('y label')

plt.title("Simple Plot")

plt.show()
```

### Here we do some other analysis to make a different graph

```python
data_arrayRand1 = np.random.randn(1000)
data_arrayRand2 = np.random.randn(1000)*0.3 + 4
combined_data = [data_arrayRand1,data_arrayRand2]
n, bins, patches = plt.hist(combined_data,50)
plt.legend(['Data 1','Data 2'])
plt.xlabel('X label')
plt.ylabel('Y label')
```
